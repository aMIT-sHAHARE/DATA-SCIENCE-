## MEAN DEVIATION

WHY ITS NEED => IF WE HAVE DATA WITH SAME MEAN THEN WHAT CHOOSE THATS WHY USE MD AND IF U FIND MD U CHOOSE SMALL MD

Mean Deviation (MD) is needed in this situation because:

Measures Spread: 
It quantifies the average distance of each data point from the mean, showing how spread out or "scattered" the data is.

Identifies Consistency: 
If you have to choose between two datasets (like the performance of two players or the quality of two product batches) that have the same average, you choose the one with the smaller MD because it indicates higher consistency and less variability. 

Small MD: Data points are closer to the mean; the dataset is more consistent.

Large MD: Data points are farther from the mean; the dataset is more inconsistent.

```python
mad_a=np.sum(np.abs(sec_a-mean))/len(sec_a)
mad_b=np.sum(np.abs(sec_b-mean))/len(sec_b)
```
