# Range

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("./titanic_clean.csv")
dataset.head(5)
```

```python
max=dataset["Age"].max()
```

```python
print(f"MAX : {max}")
```

```python
min=dataset["Age"].min()
```

```python
print(f"MIN : {min}")
```

```python
range=max-min
```

```python
print(f"Range : {range}")
```
