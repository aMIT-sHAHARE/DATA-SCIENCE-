```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("./titanic_clean.csv")
dataset.head(5)
```

### For checking null

```python
dataset.isnull().median()
```

## Find median of Age

```python
dataset["Age"].median()
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
```

```python
md=np.median(dataset["Age"])
```

```python
sns.histplot(x="Age",data=dataset,bins=[i for i in range(0,90,10)])
plt.plot([md for i in range(0,300)],[ i for i in range(0,300)],c="red")
plt.show()
```

## Find median of Fare 

```python
dataset["Fare"].median()
```

```python
md=np.median(dataset["Fare"])
```

```python
sns.histplot(x="Fare",data=dataset,bins=[i for i in range(0,90,10)])
plt.plot([md for i in range(0,300)],[ i for i in range(0,300)],c="red")
plt.show()
```

```python

sns.boxplot(x=dataset["Fare"], color="skyblue")
plt.title("Box Plot of Fare")
plt.show()
```

```python
sns.scatterplot(data=dataset, x="Age", y="Fare", alpha=0.6)
plt.title("Scatter Plot: Age vs Fare")
plt.show()
```

```python
sns.kdeplot(dataset["Fare"], fill=True, color="blue")
plt.title("Density Plot of Fare")
plt.show()
```

```python
# 1. Create the figure size
plt.figure(figsize=(10, 8))

# 2. Add numeric_only=True to filter out strings automatically
# This prevents the "could not convert string to float" error
correlation_matrix = dataset.corr(numeric_only=True)

# 3. Generate the heatmap
# fmt=".2f" ensures numbers stay readable with 2 decimal places
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Dataset Correlation Heatmap")
plt.show()
```

```python
sns.stripplot(x=dataset["Fare"], color="purple", alpha=0.5)
plt.title("Dot Plot (Strip Plot) of Fare")
plt.show()
```
