### Class and object


```python
class Car :
    def start(self):
        print(f"{self.color} {self.model} car is starting")

# creation of object and set attribute

car1=Car()
#
car1.model="Toyota"
car1.color="Red"

car2=Car()

car2.model="Honda"
car2.color="blue"

# method call 

car1.start()
car2.start()


```

#"__init__" this is a constructor use for define method and property

```python
class Car:
    def __init__(self, color, model):
        self.color = color
        self.model = model

    def start(self):
        print(f"{self.color} {self.model} car is starting")
```

```python
car1 = Car("Red", "BMW")
print(f"{car1.color} {car1.model} car is starting......")
car1.start()
```

### The __str__() method is a special method that controls what is returned when the object is printed:



```python
# without __str__() method:

class person :
    def __init__(self,name):
        self.name= name

    def show_name(self):
        print(f"THIS is me and I am {self.name}")

p1 = person("Millionaire")
p1.show_name()
```

```python
# With the __str__() method:

class person :
    def __init__(self,name):
        self.name= name

    def __str__(self):
        return f"THIS is me and I am {self.name}"

p1 = person("Millionaire")
print(p1)
```

```python
class playlist:
    def __init__(self,name):
        self.name= name
        self.songs=[]

    def add_song(self,song):
        self.songs.append(song)
        print(f"ADDED SONG :{song}")

    def remove_song(self,song):
        self.songs.remove(song)
        print(f"REMOVE SONG :{song}")
        
    def show_songs(self):
        print(f"Playlist {self.name}: ")
        for song in self.songs:
            print(f"-{song}")


my_playlist=playlist("Favorites")
my_playlist.add_song("Bohemian Rhapsody")
my_playlist.add_song("Stairway to Heaven")
my_playlist.add_song("dangle")
my_playlist.show_songs()

```

### Inheritance


```python
# polimorphism 
# static== overloding== compile type poly..
# dynamic== overriding== runtime 


class Bird:
    def fly(self):
        print("Bird is flying..")

class Airplane :
    def fly(self):
        print("Airplane is flying..")
```

```python
bird=Bird()
airplane=Airplane()
```

```python
bird.fly()
airplane.fly()
```
