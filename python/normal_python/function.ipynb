As part of your role in a data-driven project,
you need to take input from the user for their
age, weight, and height. However, there's an issue:
the user has entered their age as a negative number.
You need to correct the age of the user, and after
correcting the age, calculate their Basal Metabolic Rate (BMR)
using the Harris-Benedict equation for men.
Ensure that the final answer of BMR is an integer.

Basal Metabolic Rate (BMR):
Formula (Harris-Benedict equation for men):
BMR = 88.362 + (13.397 * weight in kg) +
(4.799 * height in cm) - (5.677 * age in years)

```python
age = int(input("Enter the your age : "))
weight = float(input("Enter the your weight in kg : "))
height = float(input("Enter the your height in cm : "))

print(f"age: {age} weight: {weight} height: {height} " )
age=abs(age)
print(f"age: {age}")

BMR=88.362 +(13.397*weight)+(4.799*height)-(5.677*age)
print(f"BMR : {int(BMR)}")
```

write a python code for hostel students, ask them to mention their course name and age. for 5 students and calculate their avg age.

```python
age=0
total_age=0
number_of_student=5
for i in range(1 ,number_of_student+1):
    print(f"\nStudent {i}")
    course=input("Enter course name : ")
    age=int(input("Enter an age : "))
    
    total_age+=age
    print(F"Course name : {course}")
    print(F"Age : {age}")
    
avg_age=total_age/number_of_student

print("\n------------------------------------------------------------")
print(f"Average age of student : {avg_age:.2f}")
```

You are working on a project to analyze weather data.
Create variables to store the current temperature,
the minimum temperature recorded, and the maximum temperature recorded.
Determine whether the current temperature is within
the range of the minimum and maximum temperatures
recorded using logical operators

```python
current_temperature=200
print(f"CURRENT TEMPERATURE : {current_temperature}")
min_temperature=5
print(f"MINIMUM TEMPERATURE : {min_temperature}")
max_temperature=40
print(f"MAXIMUM TEMPERATURE : {max_temperature}")

print("----------------------------------------------------------------------")

if(current_temperature >= min_temperature and current_temperature <= max_temperature):
    print(f"Current temperature ({current_temperature}) is within the range of the minimum and maximum temperatures")
else:
    print(f"Current temperature ({current_temperature}) is out of the range of the minimum and maximum temperatures")
    
```
